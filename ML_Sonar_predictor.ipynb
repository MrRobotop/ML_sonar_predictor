```python
#My first ever ML Project in 2019

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonar_data = pd.read_csv('data/sonar_data.csv', header=None)
```

```python
sonar_data.describe()
```

```python
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

```python
data = sonar_data.drop(columns=60, axis=1)
labels = sonar_data[60]
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size = 0.1, stratify = labels, random_state = 16)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
x_train_predictions = model.predict(x_train)
train_accuracy = accuracy_score(x_train_predictions, y_train)

print("Model Accuracy: ", train_accuracy)
```

```python
x_test_predictions = model.predict(x_test)
test_accuracy = accuracy_score(x_test_predictions, y_test)

print('Testing Accuracy: ', test_accuracy)
```
